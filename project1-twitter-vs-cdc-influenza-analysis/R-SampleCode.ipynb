```r
# PROBLEM 1: 
# Define two synthetic vectors of data representing sales over 12 months for 2 items. Compare the two using lines graphs

sales1 <- c(12,14,16,29,30,45,19,20,16,19,34,20)
sales2 <- rpois(12,34) # Generates 12 random numbers with poisson dist mean at 32
par(bg="cornsilk")
plot(sales1, col="blue", type="o", ylim=c(0,100), xlab="Month", ylab="Sales")
title(main="Sales by Month")
lines(sales2, type="o", pch=22, lty=2, col="red")
grid(nx=NA, ny=NULL)
legend("topright", inset=.05, c("Sales1", "Sales2"), fill=c("blue", "red"), horiz=TRUE)
```

```r
# PROBLEM 2:
# The sales data is available in a table in a text file. Read it in and draw a side-byside histogram to compare the performance.

sales <- read.table("data/SalesData.txt", header=T)  # Read sales data from file SalesData.txt
sales # Print to check if the data is read from the file
barplot(as.matrix(sales), main="Sales Data", ylab="Total", beside=T, col=rainbow(5))
```

```r
# PROBLEM 3
# Use boxplot to compare the two sales data.

fn<-boxplot(sales,col=c("orange","green"))$stats
text(1.45, fn[3,2], paste("Median =", fn[3,2]), adj=0, cex=.7)
text(0.45, fn[3,1],paste("Median =", fn[3,1]), adj=0, cex=.7)
grid(nx=NA, ny=NULL)
```

```r
# PROBLEM 6:
# Accessing external APIs: eg. Google map lat-long API: map command.
# The idea here is to plot the results of analysis on a map: geographical or otherwise. List a
# collection of cities you have visited and plot it on a map.

library("ggmap")
library("maptools")
library(maps)
visited <- c("SFO", "Chennai", "London", "Melbourne", "Lima,Peru", "Johannesbury, SA")
ll.visited <- geocode(visited)
visit.x <- ll.visited$lon
visit.y <- ll.visited$lat
map("world", fill=TRUE, col="white", bg="lightblue", ylim=c(-60, 90), mar=c(0,0,0,0))
points(visit.x,visit.y, col="red", pch=36)
```

```r
# PROBLEM 8
# Working with ggplot2 package (), loading a package, installing package. Object oriented
# and incremental additions (extensibility) are special features of this package. We
# can layer the commands to a base plot.

# Example of historgram on genre of books being sold by different publications. The example layers the commands on base plot
# to show the stats.

books_data <- read.csv("data/BooksSoldbyPublishers.csv", header=T) # Read books data from file BooksSoldbyPublishers.csv
head(books_data) # Print head and tail to check the data
tail(books_data)
library(ggplot2)
base <- ggplot( books_data, aes(sold.by) )
base <- base + geom_bar(aes(fill = genre), width = 0.7) +
       labs(title = "Histogram of Books Sold by Different Publications") +
       theme( axis.text.x = element_text(vjust=.45, angle=45))
base # Display
base + coord_flip()  # Now flip and display
```
